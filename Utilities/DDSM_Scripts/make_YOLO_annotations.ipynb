```python
import pandas as pd
import os
import numpy as np
from PIL import Image
import xml.etree.ElementTree as ET
from PIL import Image
from sklearn.model_selection import train_test_split
```

```python
files=os.listdir('/home/anvit/Desktop/YOLO/darknet_inbreast/data/obj/')
train,val = train_test_split(files, test_size=0.15, random_state=42)
train_file=open('/home/anvit/Desktop/YOLO/darknet_inbreast/data/train.txt','w')
val_file=open('/home/anvit/Desktop/YOLO/darknet_inbreast/data/val.txt','w')
for file in train:
    train_file.write('data/obj/'+file+'\n')
for file in val:
    val_file.write('data/obj/'+file+'\n')
for file in files:
    txtfil = open('/home/anvit/Desktop/YOLO/darknet_inbreast/data/obj/'+file[:-3] + "txt", 'w')
    xml = '/home/anvit/Desktop/YOLO/all_annotations/' + file[:-3] + "xml"
    if(os.path.exists(xml)==False):
        txtfil.close()
        continue
    tree = ET.parse(xml)
    root = tree.getroot()
    img = Image.open('/home/anvit/Desktop/YOLO/darknet_inbreast/data/obj/' + file)
    width, height = img.size
    print(width,height)
    for obj in root.findall('object'):
        typename=obj.find('name').text
        if(typename=='BENIGN_WITHOUT_CALLBACK' or typename=='BENIGN'):
            classval = 0
        elif(typename=='MALIGNANT'):
            classval = 1
        name = obj.find('bndbox')
        xmin=float(name.find('xmin').text)
        xmax=float(name.find('xmax').text)
        ymin=float(name.find('ymin').text)
        ymax=float(name.find('ymax').text)
        print(xmin,xmax,ymin,ymax)
        xcentre = (xmin + xmax)/2
        ycentre = (ymin + ymax)/2
        w = xmax - xmin
        h = ymax - ymin
        if(w==0):
            w=3
        if(h==0):
            h=3
        xcentre_norm = xcentre/width
        ycentre_norm = ycentre/height
        w_norm = w/width
        h_norm = h/height
        towrite = str(classval) + " " + str(xcentre_norm) + " " + str(ycentre_norm) + " " + str(w_norm) + " " + str(h_norm) + "\n"
        txtfil.write(towrite)
    
    txtfil.close()
```
