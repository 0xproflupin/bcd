```python
import numpy
from PIL import Image
import os
```

```python
os.chdir('/home/anvit/DDSM/bimagesjpg/')
#os.chdir('/home/anvit/Desktop/tempfold/')
count_unmatched = 0

for file in os.listdir('.'):
    im = Image.open(file)
    width, height = im.size
    print("In jpg: ", str(width), str(height))

    #Compare dimensions with .ics file
    arr = file.split('.')
    makeics = arr[0].split('_')
    icsfilename = makeics[0]+"-"+makeics[1]+"-"+makeics[2]+".ics"
    
    ics = open('/home/anvit/DDSM/bimagesics/'+icsfilename,'r')
    lines=ics.readlines()
    num_lines = len(lines)
    
    if(arr[1] == "LEFT_CC"):
        res = lines[num_lines-4].split(' ')
    elif(arr[1] == "LEFT_MLO"):
        res = lines[num_lines-3].split(' ')
    if(arr[1] == "RIGHT_CC"):
        res = lines[num_lines-2].split(' ')
    elif(arr[1] == "RIGHT_MLO"):
        res = lines[num_lines-1].split(' ')
    
    print("Actual: ",res[4],res[2])
    if(str(width) !=  res[4] or str(height) != res[2]):
            count_unmatched += 1
    
    ics.close()
```

```python
count_unmatched
```
