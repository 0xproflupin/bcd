```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import sys
import xml.etree.ElementTree as ET
from PIL import Image
from pascal_voc_writer import Writer
import math
```

```python
os.chdir('/home/anvit/Desktop/ia/annotations/')
o=0
x1=[]
x2=[]
y1=[]
y2=[]
width,height=(384,384)
for xml in os.listdir('.'):
    print(xml)
    tree = ET.parse(xml)
    root = tree.getroot()
    for bndbox in root.iter('bndbox'):     
        x1=int(bndbox[0].text)
        x2=int(bndbox[1].text)
        y1=int(bndbox[2].text)
        y2=int(bndbox[3].text)
        if(x1<1):
            x1=1
        if(x2<1):
            x2=1
        if(y1<1):
            y1=1
        if(y2<1):
            y2=1
        bndbox[0].text=str(x1)
        bndbox[1].text=str(x2)
        bndbox[2].text=str(y1)
        bndbox[3].text=str(y2)
    tree.write('/home/anvit/Desktop/ia/annotations2/'+xml)
```
