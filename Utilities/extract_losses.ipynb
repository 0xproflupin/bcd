```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
file = open("/home/anvit/py-faster-rcnn_inbreast/nohup.out", "r")
lines=file.readlines()
losses=[]
iterations=[]
df=pd.DataFrame(columns=["iteration","loss"])
for line in lines:
    ind=line.find(", loss = ")
    if(ind!=-1):
        i1=line.find("Iteration")+10
        i2=line.find("(")-1
        i3=line.find("loss = ")+7
#         print(line[i1:i2])
#         print(line[i3:-1])
        iterations.append(float(line[i1:i2]))
        losses.append(float(line[i3:-1]))
df['iteration']=np.array(iterations)
df['loss']=np.array(losses)

plt.plot(iterations,losses)
plt.figure()
plt.plot(iterations[0:100],losses[0:100])
plt.figure()
plt.plot(iterations[len(losses)-101:],losses[len(losses)-101:])
plt.show()
```

```python
file = open("/home/anvit/py-faster-rcnn_cal/calcification_expert.txt", "r")
lines=file.readlines()
losses=[]
iterations=[]
df=pd.DataFrame(columns=["iteration","loss"])
for line in lines:
    ind=line.find(", loss = ")
    if(ind!=-1):
        if(line[0]=='s'):
            line=line[7:]
        if(line[0]=='0'):
            line=line[9:]
#         print(line)
        line=line[54:]
        i1=line.find("(")-1
        i2=line.find("loss")+7
        #print(line[:i1])
        iterations.append(float(line[:i1]))
        losses.append(float(line[i2:-1]))
df['iteration']=np.array(iterations)
df['loss']=np.array(losses)

plt.plot(iterations,losses)
plt.figure()
plt.plot(iterations[0:100],losses[0:100])
plt.figure()
plt.plot(iterations[len(losses)-101:],losses[len(losses)-101:])
plt.show()
```
